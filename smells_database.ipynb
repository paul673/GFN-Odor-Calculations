```python
import os
```

```python
def read_json(filepath):
    print(f"Reading {filepath}...")
    # Open the file and read its contents
    with open(filepath, 'r') as file:
        filecontent = file.read()

    # Parse the JSON content into a dictionary
    datadict = json.loads(filecontent)
    return datadict
```

```python


# Directory path
directory = 'data'
compounds = []
# Iterate through files in the directory
for filename in os.listdir(directory):
    print(os.listdir(directory))
    if '.json' in filename:
        filepath = os.path.join(directory, filename)  # Get full path
        if os.path.isfile(filepath):  # Check if it's a file
            datadict = read_json(filepath)
            for compound in datadict['compound']:
                del compound['pivot']
            compounds + datadict['compound']
compounds
```
